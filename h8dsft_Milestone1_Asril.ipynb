### Apakah hasil produksi energi terbarukan di China mengalami kenaikan setiap tahun?

Untuk menjawab masalah tersebut maka dilakukan analisa inferential Two-Sample Independent Test

H0 : Rerata Hasil Produksi tahun 2014 tidak mengalami kenaikan signifikan terhadap tahun 2023

H1 : Rerata Hasil Produksi tahun 2014 mengalami kenaikan signifikan terhadap tahun 2023

```python
# Menampilkan isi dataframe 'data_china_yoy'
data_china_yoy
```

Isi data dari `data_china_yoy` berupa rata-rata data per tahun di China

```python
# Menampilkan total produksi per tahun dari 4 sumber energi
for i in range(2002,2023) :
    t_stat, p_val = stats.ttest_ind(data_china_yoy[data_china_yoy['Year']==i]['Production (GWh)'],data_china_yoy[data_china_yoy['Year']==(i+1)]['Production (GWh)'])
    print(f"P-Value {i}:{i+1} yaitu {p_val}")
```

Terlihat pada hasil analisa inferential terdapat banyak P-Value yang bernilai lebih dari 0.05 yang artinya H0 berhasil ditolak dan H1 berhasil diterima. Hanya di tahun 2019-2020 saja yang perbedaan hasil produksinya tidak signifikan

Dengan statement diatas dapat disimpulkan bahwa terdapat perbedaan yang signifikan dari hasil produksi energi tiap tahunnya, untuk mengetahui naik atau turunnya akan dilakukan analisa secara visual

```python
# Membuat tabel total hasil produksi dari 4 sumber energi terbarukan
total_production_yoy = data_china_yoy.groupby(['Year'])[['Production (GWh)']].sum().reset_index()

# Memvisualisasikan total hasil produksi dari 4 sumber energi per tahun
sns.lineplot(data=total_production_yoy, x='Year',y='Production (GWh)')
plt.title('Total Produksi Energi Terbarukan per Tahun')
plt.xticks(rotation=45);
```

### Berapa nilai Investasi yang dibutuhkan per tahun?

Dalam menjawab nilai investasi yang dibutuhkan/dihabiskan per tahun dapat digunakan analisa central tendency, standar deviasi, skewness, dan kurtosis untuk mencari upper boundary/batas atas

```python
# Menjumlahkan nilai investasi 5 sumber energi dalam 1 tahun
total_invest = data_china_yoy.groupby(['Year'])[['Investments (USD)']].sum().reset_index()

# Mencari Central Tendency kolom 'Investments (USD)'
print(f"Rerata Investasi per tahun : {total_invest['Investments (USD)'].mean()}")
print(f"Median Investasi per tahun : {total_invest['Investments (USD)'].median()}")
print(f"Modus Investasi per tahun : {total_invest['Investments (USD)'].mode().values[0]}")

# Mencari tahu nilai Skewness
print(f"Nilai Skewness kolom 'Investments (USD)' : {total_invest['Investments (USD)'].skew()}")
print(f"Nilai Kurtosis kolom 'Investments (USD)' : {total_invest['Investments (USD)'].kurtosis()}")
```

Hasil perhitungan Central Tendency dapat diprediksi bahwa persebaran data `Investments (USD)` merupakan distribusi normal, diperkuat dengan nilai Skewness yang berada diantara -0.5 dan 0.5, dengan mengetahui jenis distribusinya maka untuk mencari nilai batas atas dan batas bawah bisa menggunakan metode Z-Score.

```python
# Memakai confidence interval dengan confidence level 95%
std = data_china_yoy['Investments (USD)'].std()
N = len(data_china_yoy)
low, up = stats.norm.interval(
    0.95,
    loc = data_china_yoy['Investments (USD)'].mean(),
    scale = std/np.sqrt(N)
)
print(f'Nilai Investasi Limit Atas : {up}')  # Nilai Investasi Limit Atas
print(f'Nilai Investasi Limit Bawah : {low}') # Nilai Investasi Limit Bawah
```

Dari perincian analisa diatas maka dapat ditarik kesimpulan bahwa investasi yang akan masuk yaitu sebesar `$4.176.408.194,86 USD` hingga `$4.989.617.168,00 USD`
